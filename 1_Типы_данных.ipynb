```python
age = '22'              # str -> '22'    # строка
int_age = int(age)      # int -> 22      # целые числа
float_age = float(age)  # float -> 22.0  # дробные числа
bool_age = bool(age)    # bool -> True   # логический тип
print(age)
print(int_age)
print(float_age)
print(bool_age)
```
